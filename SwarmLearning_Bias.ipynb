```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulated swarm learning agents
agents = ["Agent1", "Agent2", "Agent3", "Agent4", "Agent5"]

# Lists to store fairness scores for each agent in each iteration
agent_fairness_scores = {agent: [] for agent in agents}

# Simulated fairness metric (e.g., disparate impact)
def fairness_metric(predictions, protected_group):
    # Compute fairness metric (e.g., disparate impact)
    group_positive = sum(predictions[protected_group])
    group_negative = len(predictions[protected_group]) - group_positive
    total_positive = sum(predictions)
    total_negative = len(predictions) - total_positive
    return (group_positive / total_positive) / (group_negative / total_negative)

# Simulated swarm learning iterations
num_iterations = 10
for iteration in range(num_iterations):
    print(f"Iteration {iteration + 1}:")

    # Simulate model updates and predictions for each agent
    for agent in agents:
        predictions = np.random.choice([0, 1], size=100, p=[0.7, 0.3])

        # Simulated protected group (e.g., gender)
        protected_group = np.random.choice([True, False], size=100)

        # Calculate fairness metric for this agent's predictions 
        fairness_score = fairness_metric(predictions, protected_group)

        # Store the fairness score for this agent in this iteration
        agent_fairness_scores[agent].append(fairness_score)

        # Checking for bias
        if fairness_score < 0.8:
            print(f"{agent}: Bias detected! Fairness score: {fairness_score}")
        else:
            print(f"{agent}: No bias detected. Fairness score: {fairness_score}")


# Fairness scores over iterations for each agent
for agent in agents:
    plt.plot(range(1, num_iterations + 1), agent_fairness_scores[agent], label=agent)

plt.xlabel("Iteration")
plt.ylabel("Fairness Score")
plt.title("Fairness Scores Over Swarm Learning Iterations ")
plt.legend()
plt.grid(True)
plt.show()
```
